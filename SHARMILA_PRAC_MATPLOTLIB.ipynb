```python
import matplotlib.pyplot as plt
```

```python
x= [ 2,4,6,7]
y=[10,20,15,35]
plt.plot(x,y)
```

```python
# BAR CHART
plt.bar(x,y)
plt.legend(['mild'])
```

```python
# HISTOGRAM
x=[1,4,6,7,3.4,1.5,1.2,2.7]
# interval -inf to 1, 1 to 2, 2to 3, 3 to inf
plt.hist(x,bins=[1,2,3])
# 1-2 = 1,1.5,1.2 - 3 data
# 2 to 3 = 2.7
```

```python
x=[1.5,1.7,2.2,2.5,2.8,2.6,3.5]
# interval -inf to 1, 1 to 2, 2to 3, 3 to inf
plt.hist(x,bins=[1,2,3])
```

```python
# scatter plot
wt =[ 50,55,56.7,65,68]
ht=[153,155,157,165,169]

plt.scatter(wt,ht)

plt.xlabel("weight")
plt.ylabel("height")

plt.title("HEIGHT WEIGHT CHART")
```

```python
x= [1.5 , 2.2,  3.0, 4.5,5.0]
y = [2.0,2.8,3.5,4.0,4.8]
plt.scatter(x,y)
plt.show()
```

```python
# pie chart
data = ['india','usa','uk','norway']
s=[12,4,6,8]
plt.pie(s,labels=data)
```

```python
# pie chart
data = ['india','usa','uk','norway']
s=[12,4,6,8]
ex=[0.5,0,0,0]
plt.pie(s,explode=ex, labels=data)
```

```python
# 3dimensional
three=plt.axes(projection='3d')
three.plot(s,ex)
```

```python
# scatter plot
# assign various colors - species
colors=['blue','red','green']
sps=['Iris-setosa','Iris-versicolor','Iris-virginica']

for i in range(3):
  x=df[df['Species']==sps[i]]
  plt.scatter(x['SepalLengthCm'], x['SepalWidthCm'],c=colors[i],label=sps[i])
plt.title("IRIS FLOWER ")
plt.xlabel('SEPAL LENGTH')
plt.ylabel('SEPAL WIDTH')
plt.legend()
```

```python
# scatter plot
# assign various colors - species
colors=['blue','red','green']
sps=['Iris-setosa','Iris-versicolor','Iris-virginica']

for i in range(3):
  x=df[df['Species']==sps[i]]
  plt.scatter(x['PetalLengthCm'], x['PetalWidthCm'],c=colors[i],label=sps[i])
plt.title("IRIS FLOWER ")
plt.xlabel('petal LENGTH')
plt.ylabel('petal WIDTH')
plt.legend()
```
